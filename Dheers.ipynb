```python
#Loading all jsons
import os
folder_path = '/content/drive/MyDrive/DS Project Dataset'  # Update this path to your folder

# Create an empty list to store data
data_frames = []

# Loop through all the JSON files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        # Load the JSON file
        with open(os.path.join(folder_path, filename), 'r') as file:
            data = json.load(file)

            # Normalize the data into a pandas DataFrame and append to the list
            df = pd.json_normalize(
                data['playlists'],
                record_path='tracks',
                meta=['pid', 'name', 'description', 'modified_at', 'num_artists', 'num_albums',
                      'num_tracks', 'num_followers', 'num_edits', 'duration_ms', 'collaborative'],
                meta_prefix='playlist_',  # Add prefix to avoid name conflict
                record_prefix='track_',   # Add prefix for track-specific fields
                errors='ignore'
            )
            data_frames.append(df)

# Concatenate all DataFrames
combined_df = pd.concat(data_frames, ignore_index=True)
```
